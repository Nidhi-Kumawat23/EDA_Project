```python
port_sten = PorterStemmer()
```

```python
from os.path import join
def stemming(content):

  stemmed_content =re.sub('[^a-zA-Z]',' ', content)  #it will only contain the words form a-z and A-Z no sepcial charcters or nubmers
  stemmed_content=stemmed_content.lower()
  stemmed_content =stemmed_content.split() #split the tweet in words in list
  #change the words into root words and remove stopwords
  stemmed_content =[port_sten.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)

  return stemmed_content
```

```python
twitter_data['stemmed_content'] = twitter_data['text'].apply(stemming)
```
